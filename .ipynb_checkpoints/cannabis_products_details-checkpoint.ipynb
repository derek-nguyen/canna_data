```python
import pandas as pd
import csv
from bs4 import BeautifulSoup
import requests
import IPython

# adding these modules to slow down crawl-rate
# prevents us from being banned
from time import time, sleep
from random import randint
```

# One-Product

```python
url = 'https://www.leafly.com/products/cannabis/flower?page=1'
brands = requests.get(url)
soup = BeautifulSoup(brands.content, 'html.parser')
```

### Grab first-product

```python
product_containers = soup.find_all('a',{'class':'item'})
```

```python
first_product = product_containers[0]
first_product
```

```python
# product name
first_product_name = first_product.find('div', class_ = 'product-name').text
first_product_name
```

```python
# product image url

first_image = first_product.find('img', attrs = {'class':'b-lazy'})
first_image['data-src']
```

### Grab first-product-detail

```python
url = 'https://www.leafly.com/products/details/true-humboldt-gg'
first_image['data-src']
details = requests.get(url)
soup_detail = BeautifulSoup(details.content, 'html.parser')

# product_detail_container = 
```

```python
product_detail_container = soup_detail.find_all()
```

```python
first_product_detail = product_detail_container[0]
first_product_detail.prettify
```

```python
first_product_short_description = first_product_detail.find('p', class_ = "product-short-description")
# first_product_description = first_product_detail.find('p', class_ = "product-description")
```
